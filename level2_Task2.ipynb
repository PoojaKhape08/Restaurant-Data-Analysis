Level 2 – Task 2: Cuisine Combination Analysis

Objective:
To identify the most common cuisine combinations and analyze their relationship with restaurant ratings.

```python
import pandas as pd

df = pd.read_csv("Dataset.csv")
df.head()
```

Most Common Cuisine Combinations

```python
cuisine_combo = df['Cuisines'].value_counts()
cuisine_combo.head(10)
```

combinations of Average Rating

```python
avg_rating_combo = df.groupby('Cuisines')['Aggregate rating'].mean()
avg_rating_combo.sort_values(ascending=False).head(10)
```

Conclusion:
The analysis shows that North Indian cuisine and its combinations are the most commonly served among restaurants. 
However, some less common cuisine combinations such as Italian, Deli and Hawaiian, Seafood have the highest average ratings.
This indicates that unique cuisine combinations may receive better customer satisfaction.
