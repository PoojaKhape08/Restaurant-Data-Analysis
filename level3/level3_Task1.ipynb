Level 3 – Task 1: Restaurant Reviews Analysis

Objective:
To analyze restaurant reviews and identify common patterns. The task also explores the relationship between review length and restaurant ratings.

```python
import pandas as pd

df = pd.read_csv("Dataset.csv")
df.columns
```

Rating Range Distribution

```python
df['Aggregate rating'].describe()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Aggregate rating'], df['Votes'])
plt.xlabel("Rating")
plt.ylabel("Votes")
plt.title("Votes vs Rating")
plt.show()
```

Conclusion:
The analysis shows the relationship between restaurant ratings and the number of votes. 
Restaurants with higher ratings tend to receive more votes, indicating higher customer engagement and popularity.
