Level 3 – Task 2: Votes Analysis

Objective:
To identify restaurants with the highest and lowest number of votes and understand how votes indicate restaurant popularity.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset.csv")
df.head()
```

highest votes

```python
highest_votes = df.loc[df['Votes'].idxmax()]
highest_votes
```

lowest votes

```python
lowest_votes = df.loc[df['Votes'].idxmin()]
lowest_votes
```

correlation between votes and aggregate rating

```python
correlation = df['Votes'].corr(df['Aggregate rating'])
correlation
```

Conclusion:
The votes analysis identified the most and least popular restaurants based on customer engagement.
The restaurant "Toit" in Bangalore has the highest number of votes, indicating strong popularity, while some restaurants have zero votes and ratings.
The correlation between votes and aggregate rating is approximately 0.31, which indicates a moderate positive relationship. 
This suggests that restaurants with higher ratings tend to receive more votes, but the relationship is not very strong.
