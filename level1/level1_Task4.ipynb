Level 1 – Task 4: Online Delivery Analysis

Objective:
To analyze the percentage of restaurants offering online delivery and compare ratings between restaurants that offer online delivery and those that do not.

```python
import pandas as pd

df = pd.read_csv("Dataset.csv")
df.head()
```

```python
delivery_count = df['Has Online delivery'].value_counts()
percentage = (delivery_count / len(df)) * 100
percentage
```

```python
avg_rating_delivery = df.groupby('Has Online delivery')['Aggregate rating'].mean()
avg_rating_delivery
```

```python
avg_rating_delivery.plot(kind='bar')
```

Conclusion:
The analysis shows that only a smaller percentage of restaurants provide online delivery services. However, restaurants offering online delivery have a higher average rating compared to those that do not offer delivery services. This indicates that online delivery may contribute to better customer satisfaction.
