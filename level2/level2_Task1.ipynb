Level 2 – Task 1: Restaurant Ratings Analysis

Objective:
To analyze the distribution of restaurant ratings and calculate the average number of votes received by restaurants.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset.csv")
df.head()
```

Ratings Distribution

```python
rating_distribution = df['Aggregate rating'].value_counts().sort_index()
rating_distribution
```

Ratings Graph

```python
rating_distribution.plot(kind='bar')
plt.title("Ratings Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.show()
```

Average Votes Calculate

```python
average_votes = df['Votes'].mean()
average_votes
```

Conclusion:
The analysis of restaurant ratings shows that most restaurants fall within the average rating range.
A large number of restaurants have a rating of 0.0, which may indicate missing or unreviewed data. 
    
The average number of votes received by restaurants is approximately 157, indicating a moderate level of customer engagement.
