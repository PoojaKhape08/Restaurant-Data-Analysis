Level 2 – Task 3: Geographic Analysis

Objective:
To plot restaurant locations using latitude and longitude and identify areas with high restaurant density.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset.csv")
df.head()
```

Locations Scatter Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Longitude'], df['Latitude'])
plt.title("Restaurant Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

Conclusion:
The geographic analysis shows the distribution of restaurants based on latitude and longitude.
The scatter plot indicates that restaurants are concentrated in specific areas, suggesting higher restaurant density in major cities.
